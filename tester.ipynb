```python
import pandas as pd
import json

def process_predictions(answer):
    data = []
    for item in answer:
        for choice in item['choices']:
            # Parse the JSON content inside 'message' -> 'content'
            content = json.loads(choice['message']['content'])
            
            # Extract the required information
            index = choice['index']
            earnings_direction = content.get('earnings direction', '')
            magnitude = content.get('magnitude', '')
            confidence_score = content.get('confidence score', '')
            summary_of_rationale = content.get('summary of rationale', '')
            
            # Append the extracted information as a tuple to the data list
            data.append((index, earnings_direction, magnitude, confidence_score, summary_of_rationale))

    # Create a DataFrame from the extracted data
    df = pd.DataFrame(data, columns=['Index', 'Earnings Direction', 'Magnitude', 'Confidence Score', 'Summary of Rationale'])
    
    return df

# Example usage:
# Assuming 'answer' is your list of dictionaries as described
df = process_predictions(answer)
print(df)
```
